### 3. DATA CLEANING

```python
df_clean = df.copy()
```

##### 1. Mangubah type data

```python
date_columns=[
'order_purchase_timestamp',
'order_approved_at',
'order_delivered_timestamp',
'order_estimated_delivery_date'
]

for col in date_columns:
    df_clean[col]=pd.to_datetime(df_clean[col])
```

##### 3. Mengecek dan menangani missing value setelah data di drop

```python
df_clean.isnull().sum()
```

```python
df_clean[df_clean["order_delivered_timestamp"].isna()]["order_status"].value_counts()
```

INTERPRETASI :
Terdapat 81 (2.23%) data yang missing pada order_delivered_timestamp memang masuk akal.
Karena:
- Shipped → Barang sudah dikirim, tapi belum ada tanggal diterima.
- Processing → Pesanan masih diproses, jadi belum mungkin ada tanggal diterima.
- Invoiced → Baru tahap penagihan, jelas belum terkirim.
- Canceled → Pesanan dibatalkan, sehingga memang tidak akan memiliki tanggal pengiriman.
Jadi missing value ini bukan error, melainkan karakteristik proses bisnis.

##### Missing value kolom order_delivered_timestamp

- order_delivered_timestamp tidak dilakukan imputasi
- karena missing value berasal dari order yang belum selesai atau dibatalkan.

##### Mengisi missing value product_category_name dengan "Unknown

```python
df_clean["product_category_name"] = df_clean["product_category_name"].fillna("Unknown")
```

##### Mengisi missing value pada beberapa kolom berikut dengan nilai "Median"

```python
cols=[
'product_weight_g',
'product_length_cm',
'product_height_cm',
'product_width_cm'
]

for col in cols:
    df_clean[col]=df_clean[col].fillna(df_clean[col].median())
```

### 4. Data Manipulation

##### FEATURING ENGINEERING

##### 1. Membuat kolom Revenue

```python
df_clean["Revenue"]=df_clean["payment_value"]
```

##### 2. Membuat kolom Delivery Days

```python
# Order yang belum memiliki tanggal pengiriman akan bernilai NaN

df_clean["Delivery Days"] = (
    df_clean["order_delivered_timestamp"] -
    df_clean["order_purchase_timestamp"]
).dt.days
```

```python
df_clean[[
    "order_purchase_timestamp",
    "order_delivered_timestamp",
    "Delivery Days"
]].head()
```

##### 3. Membuat kolom Purchase Year

```python
df_clean["Purchase Year"] = (df_clean["order_purchase_timestamp"].dt.year)
```

##### 4. Membuat kolom Purchase Quarter

```python
df_clean["Purchase Quarter"] = (df_clean["order_purchase_timestamp"].dt.quarter)
```

##### 5. Membuat kolom Purchase Month

```python
df_clean["Purchase Month"] = (df_clean["order_purchase_timestamp"].dt.month_name())
```

##### 6. Membuat kolom Purchase Year-Month

```python
df_clean["Year-Month"] = (df_clean["order_purchase_timestamp"].dt.to_period("M").astype(str))
```

##### 7. Membuat Shipping Ratio

```python
# Mengukur proporsi biaya pengiriman terhadap harga produk
df_clean["Shipping Ratio"] = (df_clean["shipping_charges"] /df_clean["price"])
```

##### 8. Membuat kolom Product Volume

```python
# Karena satu baris merepresentasikan satu transaksi produk, kita cukup membuat indikator.
df_clean["Product Sold"] = 1
```

##### 9. Membuat kolom Delivery Status

```python
df_clean["Delivery Status"] = np.where(
    df_clean["order_delivered_timestamp"].isna(),
    "Not Delivered",
    "Delivered"
)
```

##### 10. Membuat kolom Delivery Performance

```python
df_clean["Delivery Performance"] = np.where(
    df_clean["order_delivered_timestamp"] <= df_clean["order_estimated_delivery_date"],
    "On Time",
    "Late"
)
df_clean.loc[
    df_clean["order_delivered_timestamp"].isna(),
    "Delivery Performance"
] = "Not Delivered"
```

##### 11. Membuat kolom Repeat Customer

```python
customer_order = (
    df_clean.groupby("customer_id")["order_id"]
    .nunique()
)

df_clean["Repeat Customer"] = (
    df_clean["customer_id"]
    .map(customer_order)
)
df_clean["Repeat Customer"] = np.where(
    df_clean["Repeat Customer"] > 1,
    "Repeat",
    "One Time"
)
```

##### 12. Membuat kolom Product Volume

```python
df_clean["Product Volume"] = (
    df_clean["product_length_cm"] *
    df_clean["product_width_cm"] *
    df_clean["product_height_cm"]
)
```

##### 3. MULTIVARIATE ANALYSIS

##### a. Product Category × Payment Type

Untuk melihat apakah kategori produk tertentu memiliki kecenderungan menggunakan metode pembayaran tertentu

INTERPRETASI :
- Beberapa kategori produk menunjukkan preferensi metode pembayaran tertentu, dengan Credit Card mendominasi hampir seluruh kategori.

##### b. Product Category × Order Status

Untuk melihat apakah kategori tertentu memiliki tingkat cancle atau processing yang lebih tinggi

```python
status_category = pd.crosstab(df_clean["product_category_name"],df_clean["order_status"])
status_category = status_category.sort_values(by="delivered",ascending=False).head(10)
status_category.plot(kind="bar",stacked=True,figsize=(12,6))

plt.title("Order Status by Product Category", fontsize=14, fontweight='bold')
plt.xlabel("Product Category", fontsize=10, fontweight='bold')
plt.ylabel("Number of Orders", fontsize=10, fontweight='bold')
plt.xticks(rotation=45, ha="right")

plt.tight_layout()
plt.show()
```

INTERPRETASI :
- Mayoritas kategori memiliki tingkat penyelesaian pesanan (delivered) yang tinggi, sementara status canceled dan processing hanya muncul pada sebagian kecil kategori. Bahkan untuk top 10 product category ini terlihat sangat sedikit order statusnya yang tidak delivered
